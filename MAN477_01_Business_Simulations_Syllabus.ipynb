###**Course Definition**
Teaching to run a manufacture company via a game (CapsimCore) based on the simulating the dynamics of a competitive environment.
<br><br>
###**Course Objectives**
The aim of this course is to provide the most concrete realization of theoretical courses taken in the first three years of business education in an online simulation environment. With the help of this course, students are expected to recognize the complexity of business life caused by internal and external factors and to understand how to cope with this complexity and how to be successful as a team member.
<br><br>
###**Learning Outcomes**
The students who succeeded in this course;

* Understand the complex structure of the business environment,
* Know the factors that measure the success of an enterprise,
* Understand the importance of internal and external factors on decisions,
* Form an awareness on the knowledge and skills required to become a strong team player.<br><br>


###**Teaching Strategies**
* Team effort, 
* Case studies,
* Debate<br><br>

###**Evaluation System**
Team efforts will be the basis of the final grading. In detail, in-group peer evalutions, reports on justification of business decisions, and game performance will be evaluated while grading individuals.
* For each decision category, students are to prepare reports on their "6 thinking hats" roles and justifying each decision. These reports will be written in standard fonts&size with 1.5 line spacing. **(Constitutes 80% of final grades)** 
* Game performance for each period will be measured with a 5-star metric. **(Constitutes 20% of final grades)**<br><br>

**P.S.:** Worst game performance of each group will not be taken into consideration in final grading.<br><br>
